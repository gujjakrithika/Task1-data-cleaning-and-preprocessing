```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/refs/heads/master/netflix_titles.csv')
df
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
```

```python
df = df.drop_duplicates()
```

```python
df['type'] = df['type'].str.lower().str.strip()  # Convert to lowercase
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.dtypes
# Convert types if needed
# df['some_column'] = df['some_column'].astype(int)
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```
